```python
import requests
from bs4 import BeautifulSoup
import math

url = "https://www.fintelligence.ir/Blog/GetBlogSearchResult"

base_payload = {
    "PageSize": "12",
    "PageNo": 1,
    "SearchKeyword": "",
    "SelectedBlogCategoryIDs": [],
    "SelectedAgeRangeIDs": [],
    "SelectedBlogContentTypeIDs": [],
    "SelectedBlogTagIDs": [],
    "SelectedBlogPostSortTypeID": "1"
}

session = requests.Session()
response = session.post(url, json=base_payload, timeout=30)
response.raise_for_status()
data = response.json()
page_count = data["pageCount"]

all_links = []

for page_no in range(1, page_count + 1):
    payload = dict(base_payload)
    payload["PageNo"] = page_no
    
    r = session.post(url, json=payload, timeout=30)
    r.raise_for_status()
    html_content = r.json().get("html", "")
    soup = BeautifulSoup(html_content, "html.parser")
    
    for a in soup.select("a.blog-post-list-item-readmore"):
        href = a.get("href")
        if href:
            all_links.append(href)

all_links = sorted(set(all_links))

print("Total unique links:", len(all_links))
for link in all_links:
    print(link)
```

```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

BASE_URL = "https://www.fintelligence.ir"
API_URL = "https://www.fintelligence.ir/Blog/GetBlogSearchResult"
AUTHOR_NAME = "کمیل رودی"

payload = {
    "PageSize": "12",
    "PageNo": 1,
    "SearchKeyword": "",
    "SelectedBlogCategoryIDs": [],
    "SelectedAgeRangeIDs": [],
    "SelectedBlogContentTypeIDs": [],
    "SelectedBlogTagIDs": [],
    "SelectedBlogPostSortTypeID": "1"
}

session = requests.Session()

# گرفتن تعداد صفحات
r = session.post(API_URL, json=payload)
data = r.json()
page_count = data["pageCount"]

dr_roudi_links = []

for page in range(1, page_count + 1):

    payload["PageNo"] = page
    r = session.post(API_URL, json=payload)
    html = r.json()["html"]

    soup = BeautifulSoup(html, "html.parser")

    links = [a["href"] for a in soup.select("a.blog-post-list-item-readmore")]

    for link in tqdm(links, leave=False):

        full_url = BASE_URL + link
        article = session.get(full_url)

        if AUTHOR_NAME in article.text:
            dr_roudi_links.append(full_url)

print("Total Dr Roudi articles:", len(dr_roudi_links))

# ذخیره لینک‌ها
with open("dr_roudi_links.txt", "w", encoding="utf-8") as f:
    for l in dr_roudi_links:
        f.write(l + "\n")
```
