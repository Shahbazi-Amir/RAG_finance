```python
import requests
from bs4 import BeautifulSoup
from tqdm import tqdm

BASE_URL = "https://www.fintelligence.ir"
API_URL = "https://www.fintelligence.ir/Blog/GetBlogSearchResult"
AUTHOR_NAME = "کمیل رودی"

payload = {
    "PageSize": "12",
    "PageNo": 1,
    "SearchKeyword": "",
    "SelectedBlogCategoryIDs": [],
    "SelectedAgeRangeIDs": [],
    "SelectedBlogContentTypeIDs": [],
    "SelectedBlogTagIDs": [],
    "SelectedBlogPostSortTypeID": "1"
}

session = requests.Session()

# گرفتن تعداد صفحات
r = session.post(API_URL, json=payload)
data = r.json()
page_count = data["pageCount"]

dr_roudi_links = []

for page in range(1, page_count + 1):

    payload["PageNo"] = page
    r = session.post(API_URL, json=payload)
    html = r.json()["html"]

    soup = BeautifulSoup(html, "html.parser")

    links = [a["href"] for a in soup.select("a.blog-post-list-item-readmore")]

    for link in tqdm(links, leave=False):

        full_url = BASE_URL + link
        article = session.get(full_url)

        if AUTHOR_NAME in article.text:
            dr_roudi_links.append(full_url)

print("Total Dr Roudi articles:", len(dr_roudi_links))

# ذخیره لینک‌ها
with open("dr_roudi_links.txt", "w", encoding="utf-8") as f:
    for l in dr_roudi_links:
        f.write(l + "\n")
```

```python
import json
from pathlib import Path

import requests
from bs4 import BeautifulSoup

BASE_URL = "https://www.fintelligence.ir"
SEARCH_URL = f"{BASE_URL}/Blog/GetBlogSearchResult"

SAVE_DIR = Path("data/raw/webpages")
SAVE_DIR.mkdir(parents=True, exist_ok=True)

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 Chrome/137.0 Safari/537.36"
    ),
    "Content-Type": "application/json;charset=UTF-8",
    "Referer": f"{BASE_URL}/Blog/List",
    "Origin": BASE_URL,
    "X-Requested-With": "XMLHttpRequest",
}

PAYLOAD = {
    "PageSize": "12",
    "PageNo": 1,
    "SearchKeyword": "",
    "SelectedBlogCategoryIDs": [],
    "SelectedAgeRangeIDs": [],
    "SelectedBlogContentTypeIDs": [],
    "SelectedBlogTagIDs": [],
    "SelectedBlogPostSortTypeID": "1",
}


def get_links(page_no: int):
    payload = PAYLOAD.copy()
    payload["PageNo"] = page_no

    response = requests.post(
        SEARCH_URL,
        headers=HEADERS,
        json=payload,
        timeout=30,
    )

    response.raise_for_status()

    data = response.json()

    html = data["html"]

    soup = BeautifulSoup(html, "html.parser")

    links = []

    for a in soup.select("a[href*='/Blog/Detail/']"):
        url = BASE_URL + a["href"]

        if url not in links:
            links.append(url)

    return links


all_links = []

page = 1

while True:
    print(f"Reading page {page}...")

    links = get_links(page)

    if not links:
        break

    all_links.extend(links)

    page += 1

all_links = sorted(set(all_links))

with open(SAVE_DIR / "links.txt", "w", encoding="utf-8") as f:
    for link in all_links:
        f.write(link + "\n")

print(f"\nTotal links: {len(all_links)}")
print("Saved to:", SAVE_DIR / "links.txt")
```

```python
import time
from pathlib import Path

import requests
from bs4 import BeautifulSoup

BASE_URL = "https://www.fintelligence.ir"

AUTHOR_ID = 4

OUTPUT_DIR = Path("data/raw/webpages")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

HEADERS = {
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 Chrome/137.0 Safari/537.36"
    )
}


def get_page(page: int) -> str:
    url = f"{BASE_URL}/Author/GetAuthorBlogByPage/{page}?auid={AUTHOR_ID}"

    response = requests.get(
        url,
        headers=HEADERS,
        timeout=30,
    )

    response.raise_for_status()

    return response.text


def extract_links(html: str):
    soup = BeautifulSoup(html, "html.parser")

    links = []

    for a in soup.select('a[href*="/Blog/Detail/"]'):
        href = a["href"]

        if href.startswith("/"):
            href = BASE_URL + href

        if href not in links:
            links.append(href)

    return links


def crawl_author():

    page = 1
    all_links = []

    while True:

        print(f"Reading page {page}")

        html = get_page(page)

        links = extract_links(html)

        if len(links) == 0:
            break

        print(f"  {len(links)} links")

        all_links.extend(links)

        page += 1

        time.sleep(0.3)

    all_links = sorted(set(all_links))

    output = OUTPUT_DIR / "kamil_roodi_links.txt"

    with open(output, "w", encoding="utf-8") as f:
        for link in all_links:
            f.write(link + "\n")

    print()
    print("=" * 40)
    print("Finished")
    print(f"Total articles : {len(all_links)}")
    print(f"Saved to       : {output}")


if __name__ == "__main__":
    crawl_author()
```

```python
from pathlib import Path
import json
import time

import requests
from bs4 import BeautifulSoup
from tqdm import tqdm


BASE_URL = "https://www.fintelligence.ir"

LINKS_FILE = Path("data/raw/webpages/kamil_roodi_links.txt")

OUTPUT_DIR = Path("data/raw/webpages/articles")

OUTPUT_DIR.mkdir(parents=True, exist_ok=True)


session = requests.Session()

session.headers.update({
    "User-Agent":
    "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 Chrome/137.0 Safari/537.36"
})


def article_id(url: str):

    return url.split("/Blog/Detail/")[1].split("/")[0]


def clean_text(text):

    return " ".join(text.split())


def download(url):

    r = session.get(url, timeout=30)

    r.raise_for_status()

    return r.text


def parse(html, url):

    soup = BeautifulSoup(html, "html.parser")

    title = ""

    if soup.title:
        title = clean_text(soup.title.text)

    body = ""

    article = (
        soup.find("article")
        or soup.find(class_="blog-detail")
        or soup.find(class_="blog-post-detail")
        or soup.find("main")
    )

    if article:
        body = clean_text(article.get_text(" ", strip=True))
    else:
        body = clean_text(soup.get_text(" ", strip=True))

    return {
        "url": url,
        "title": title,
        "content": body,
    }


def save(article, html):

    aid = article_id(article["url"])

    folder = OUTPUT_DIR / aid

    folder.mkdir(exist_ok=True)

    (folder / "article.html").write_text(
        html,
        encoding="utf-8"
    )

    (folder / "article.json").write_text(
        json.dumps(article,
                   ensure_ascii=False,
                   indent=2),
        encoding="utf-8"
    )

    md = f"# {article['title']}\n\n{article['content']}"

    (folder / "article.md").write_text(
        md,
        encoding="utf-8"
    )


def main():

    links = LINKS_FILE.read_text(
        encoding="utf-8"
    ).splitlines()

    failed = []

    for url in tqdm(links):

        try:

            html = download(url)

            article = parse(html, url)

            save(article, html)

            time.sleep(0.2)

        except Exception as e:

            failed.append({
                "url": url,
                "error": str(e)
            })

    if failed:

        with open(
            OUTPUT_DIR / "failed.json",
            "w",
            encoding="utf-8"
        ) as f:

            json.dump(
                failed,
                f,
                ensure_ascii=False,
                indent=2
            )

    print()

    print("=" * 40)

    print(f"Downloaded : {len(links)-len(failed)}")

    print(f"Failed     : {len(failed)}")


if __name__ == "__main__":
    main()
```

```python
from pathlib import Path
import json
import time

import requests
from bs4 import BeautifulSoup
from tqdm import tqdm


BASE_URL = "https://www.fintelligence.ir"

LINKS_FILE = Path("data/raw/webpages/kamil_roodi_links.txt")
OUTPUT_DIR = Path("data/raw/webpages/articles")
OUTPUT_DIR.mkdir(parents=True, exist_ok=True)

session = requests.Session()
session.headers.update({
    "User-Agent": (
        "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
        "AppleWebKit/537.36 Chrome/137.0 Safari/537.36"
    )
})


def article_id(url):
    return url.split("/Blog/Detail/")[1].split("/")[0]


def clean_text(text):
    return " ".join(text.split())


def download(url):
    r = session.get(url, timeout=30)
    r.raise_for_status()
    return r.text


def parse(html, url):

    soup = BeautifulSoup(html, "html.parser")

    title = ""
    if soup.title:
        title = clean_text(soup.title.get_text())

    content_div = soup.select_one(".blog-post-body-container .mt-md")

    body = ""

    if content_div:

        paragraphs = []

        for p in content_div.find_all("p"):

            text = clean_text(p.get_text(" ", strip=True))

            if len(text) > 20:
                paragraphs.append(text)

        # حذف پاراگراف‌های تکراری
        paragraphs = list(dict.fromkeys(paragraphs))

        body = "\n\n".join(paragraphs)

    return {
        "url": url,
        "title": title,
        "content": body,
    }


def save(article, html):

    aid = article_id(article["url"])

    folder = OUTPUT_DIR / aid
    folder.mkdir(parents=True, exist_ok=True)

    (folder / "article.html").write_text(
        html,
        encoding="utf-8"
    )

    (folder / "article.json").write_text(
        json.dumps(
            article,
            ensure_ascii=False,
            indent=2
        ),
        encoding="utf-8"
    )

    md = f"# {article['title']}\n\n{article['content']}"

    (folder / "article.md").write_text(
        md,
        encoding="utf-8"
    )


def main():

    links = LINKS_FILE.read_text(
        encoding="utf-8"
    ).splitlines()

    failed = []

    for url in tqdm(links):

        try:

            html = download(url)

            article = parse(html, url)

            save(article, html)

            time.sleep(0.2)

        except Exception as e:

            failed.append({
                "url": url,
                "error": str(e)
            })

    print("=" * 40)
    print(f"Downloaded: {len(links)-len(failed)}")
    print(f"Failed: {len(failed)}")

    if failed:
        (OUTPUT_DIR / "failed.json").write_text(
            json.dumps(
                failed,
                ensure_ascii=False,
                indent=2
            ),
            encoding="utf-8"
        )


main()
```
