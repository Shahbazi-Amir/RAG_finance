نام پروژه:
Financial Knowledge AI Assistant

نام فایل نوت‌بوک پیشنهادی:
financial_rag_assistant.ipynb

توضیح پروژه:

این پروژه یک سیستم دستیار هوش مصنوعی برای پاسخ‌گویی به سوالات در حوزه سواد مالی و هوش مالی است. این سیستم بر اساس مقالات یک وب‌سایت و مجموعه‌ای از کتاب‌های مالی ساخته می‌شود و کاربران می‌توانند سوالات خود را درباره مفاهیم مالی بپرسند و پاسخ دقیق دریافت کنند.

هسته این پروژه بر پایه معماری RAG (Retrieval Augmented Generation) طراحی می‌شود. در این معماری، ابتدا محتوای منابع (مقالات و کتاب‌ها) استخراج و پردازش می‌شود، سپس متن‌ها به بخش‌های کوچک‌تر تقسیم می‌شوند و برای هر بخش embedding تولید می‌شود. این embeddingها در یک پایگاه داده برداری (Vector Database) ذخیره می‌شوند تا امکان جستجوی معنایی فراهم شود.

زمانی که کاربر سوالی می‌پرسد، سیستم ابتدا سوال را به embedding تبدیل می‌کند و سپس مرتبط‌ترین بخش‌های متن را از پایگاه داده پیدا می‌کند. در مرحله بعد این اطلاعات به یک مدل زبانی (LLM) داده می‌شود تا پاسخ نهایی بر اساس منابع موجود تولید شود.

هدف این سیستم این است که پاسخ‌ها فقط بر اساس منابع مشخص شده تولید شوند تا از تولید پاسخ‌های نادرست یا ساختگی جلوگیری شود.

ویژگی‌های اصلی سیستم:

- استخراج و پردازش داده از مقالات وب‌سایت
- استخراج متن از کتاب‌های مالی
- پاکسازی و آماده‌سازی داده‌ها
- تقسیم متن به قطعات کوچک (Chunking)
- تولید Embedding برای جستجوی معنایی
- ذخیره داده‌ها در Vector Database
- جستجوی معنایی در میان منابع
- تولید پاسخ با استفاده از مدل زبانی
- نمایش منبع پاسخ در صورت امکان
- امکان پاسخ‌گویی به کاربران درباره مفاهیم مالی

کاربرد سیستم:

این سیستم می‌تواند به عنوان یک دستیار آموزشی برای کاربران یک وب‌سایت مالی استفاده شود تا بتوانند درباره موضوعاتی مثل سرمایه‌گذاری، مدیریت پول، استقلال مالی و اصول سواد مالی سوال بپرسند و پاسخ‌های مبتنی بر منابع معتبر دریافت کنند.

مقیاس پروژه:

سیستم به گونه‌ای طراحی می‌شود که بتواند حدود ۱۵۰۰ کاربر را پشتیبانی کند و در آینده قابلیت توسعه و اضافه شدن منابع جدید را داشته باشد.

هدف نهایی پروژه:

ساخت یک پایگاه دانش مالی هوشمند که بتواند دانش موجود در مقالات و کتاب‌های مالی را به یک دستیار تعاملی تبدیل کند و دسترسی کاربران به این دانش را ساده‌تر و سریع‌تر نماید.

تقسیم‌بندی مراحل پروژه
Financial Knowledge AI Assistant

Phase 1 — راه‌اندازی پروژه (Project Setup)
هدف: آماده‌سازی محیط توسعه.
کارها:
- ساخت ریپازیتوری پروژه
- ایجاد نوت‌بوک financial_rag_assistant.ipynb
- نصب کتابخانه‌ها (llama-index, qdrant-client, sentence-transformers, openai, pypdf, bs4)
- تنظیم API Key ها
- ساختاردهی پوشه‌های پروژه

Phase 2 — جمع‌آوری داده‌ها (Data Ingestion)
هدف: جمع‌آوری منابع دانش.
منابع:
- مقالات وب‌سایت
- کتاب‌های مالی (PDF)
کارها:
- استخراج متن از مقالات
- استخراج متن از PDF
- ذخیره داده‌ها در پوشه data/raw

Phase 3 — پاکسازی و آماده‌سازی داده (Preprocessing)
هدف: آماده‌سازی متن برای پردازش.
کارها:
- حذف HTML و نویز
- نرمال‌سازی متن
- یکسان‌سازی ساختار داده
- ذخیره در فرمت ساختاریافته (JSON)

Phase 4 — تقسیم متن (Chunking)
هدف: شکستن متن‌های طولانی به بخش‌های کوچک.
تنظیمات پیشنهادی:
- اندازه chunk حدود 500 توکن
- overlap حدود 50 تا 100 توکن
خروجی: مجموعه‌ای از قطعات متن قابل پردازش

Phase 5 — تولید Embedding
هدف: تبدیل هر chunk به بردار معنایی.
مدل‌های پیشنهادی:
- BGE-M3
- text-embedding-3-large
خروجی: embedding برای هر قطعه متن

Phase 6 — پایگاه داده برداری (Vector Database)
هدف: ذخیره embeddingها برای جستجوی معنایی.
ابزار:
- Qdrant
کارها:
- ساخت collection
- ذخیره embeddingها
- ذخیره metadata مثل عنوان، منبع و متن

Phase 7 — سیستم بازیابی (Retrieval)
هدف: پیدا کردن مرتبط‌ترین متن‌ها با سوال کاربر.
مراحل:
- تبدیل سوال به embedding
- جستجو در Vector Database
- بازیابی top-k chunk مرتبط

Phase 8 — بهبود نتایج (Reranking)
هدف: افزایش دقت نتایج بازیابی.
ابزار پیشنهادی:
- BGE Reranker
روش:
- دریافت 10 chunk
- انتخاب 3 تا 5 مورد بهترین

Phase 9 — تولید پاسخ با LLM
هدف: تولید پاسخ نهایی برای کاربر.
مدل پیشنهادی:
- GPT-4.1 یا GPT-4o
روش:
- ارسال سوال کاربر
- ارسال context بازیابی‌شده
- تولید پاسخ مبتنی بر منابع

Phase 10 — نمایش منبع پاسخ
هدف: افزایش اعتمادپذیری پاسخ‌ها.
کارها:
- نمایش منبع مقاله یا کتاب
- لینک یا عنوان منبع در پاسخ

Phase 11 — ارزیابی سیستم
هدف: بررسی کیفیت سیستم.
روش:
- طراحی 50 تا 100 سوال تست
- بررسی دقت پاسخ
- بررسی ارتباط پاسخ با منابع

Phase 12 — استقرار سیستم (اختیاری)
در صورت نیاز:
- ساخت API با FastAPI
- اتصال به وب‌سایت
- ایجاد رابط چت برای کاربران

معماری کلی سیستم:

Cleaning & Preprocessing
↓
Chunking
↓
Embedding
↓
Vector Database (Qdrant)
↓
Retrieval
↓
Reranking
↓
LLM
↓
Answer + Sources
